```python
IF_IFS = pd.DataFrame(data["IF-IFS"])
```

```python
IF_IFS
```

```python
def muturity(x):
    if x > 30:
        return 'Old'
    else:
        return 'Young'
    

IF_IFS['IF'] = IF_IFS['Age'].apply(lambda x: muturity(x))
```

```python
IF_IFS
```

```python
IF_IFS.dtypes
```

```python
for i in range(len(IF_IFS)):
    print(IF_IFS.loc[i, 'JobTitle'])
```

```python
for i in range(len(IF_IFS)):
    if IF_IFS.loc[i, 'JobTitle'] == 'Salesman':
        IF_IFS.loc[i, 'IFS'] = 'Sales'
    elif IF_IFS.loc[i, 'JobTitle'] == 'HR':
        IF_IFS.loc[i, 'IFS'] = 'Fire Immediately'
    elif IF_IFS.loc[i, 'JobTitle'] == 'Regional Manager':
        IF_IFS.loc[i, 'IFS'] = 'Give Bonus'
```
